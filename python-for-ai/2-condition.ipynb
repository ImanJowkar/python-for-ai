```python
import numpy as np
import matplotlib.pyplot as plt

# Given constant
t = 0
# Range of x values
x = np.linspace(-5, 5, 400)

# Piecewise function
y = np.where(x <= t, 0, x - t)

# Plot
plt.figure()
plt.plot(x, y)
plt.title("y = 0 if x <= t else x - t (t = 2)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
def y(x,t1,t2):
    if x < t1:
        return 13
    elif t1 <= x < t2:
        return x - 6
    elif x >= t2:
        return -9 * x

t1 , t2, x = -4, 10, np.linspace(-5, 5, 400)
output = y(x, t1, t2)
print()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def y_func(x, t1, t2):
    return np.where(
        x < t1, 13,
        np.where((x >= t1) & (x < t2), x - 6, -9 * x)
    )

t1, t2 = -4, 10
x = np.linspace(-10, 10, 400)
y = y_func(x, t1, t2)

plt.figure()
plt.plot(x, y)
plt.title("Piecewise Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```
