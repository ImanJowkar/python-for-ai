### if you want to copy a list you have to use copy method

```python
lst6 = [1,2,3]
lst7 = lst6

print(lst6)
print(lst7)
```

```python
lst7[2] = 55
print(lst6)
print(lst7)
```

```python
lst8 = [1,2,3]
lst9 = lst8.copy()

print(lst8)
print(lst9)
```

```python
lst8[2] = 55
print(lst8)
print(lst9)
```
