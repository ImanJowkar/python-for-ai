# Example

```python
scores = {'iman': [2,2,3], 'jack': [5,5,5], 'ali': [34,3,2], 'hossein': [20,20,19]}

res_dict = dict.fromkeys(scores.keys(), 0)
print(res_dict)
```

# calculate the mean score of each student

```python
for k,v in scores.items():
    res_dict[k] = sum(v) / len(v)

print(res_dict)
```

# count

```python
mylst = [1,2,3,4,3,2,1,2,3,4,2,3,2,3,4, 6, 5,5 ]
print(mylst)
```

```python
set(mylst)
```

```python
list(set(mylst))
```

```python
mydict = dict.fromkeys(list(set(mylst)), 0)
print(mydict)
```

```python
for i in list(set(mylst)):
    mydict[i] = mylst.count(i)
print(mydict)
```
