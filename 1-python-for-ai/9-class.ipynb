```python
class Router:
    def __init__(self, name, ip):
        self.name = name      # attribute
        self.ip = ip          # attribute

    def show_config(self):
        return f"Router {self.name} has IP {self.ip}"
```

```python
router1 = Router("R1", "192.168.1.1")
router1.show_config()
```

```python
class Device:
    def __init__(self, hostname, vendor):
        self.hostname = hostname
        self.vendor = vendor

    def info(self):
        return f"{self.hostname} is a {self.vendor} device"

    def connect(self):
        return f"Connecting to {self.hostname}..."

d1 = Device("CoreSwitch", "Cisco")
print(d1.info())
print(d1.connect())
```

```python
class Router(Device):
    def __init__(self, hostname, vendor, routing_protocol):
        super().__init__(hostname, vendor)
        self.routing_protocol = routing_protocol

    def protocol(self):
        return f"Routing protocol: {self.routing_protocol}"

r1 = Router("EdgeRouter", "MikroTik", "OSPF")
print(r1.info())
print(r1.protocol())
```
