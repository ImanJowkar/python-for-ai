# Tuple

```python
tup = (1,2,3,4)
print(tup)
```

```python
print(type(tup))
```

```python
for i in tup:
    print(i)
```

# unpack a tuple

```python
a1,a2,a3,a4 = tup
print(a1)
print(a2)
print(a3)
print(a4)
```

```python
a, *lst = tup
print(tup)
print(a)
print(lst)
```

```python
*lst, a = tup
print(tup)
print(a)
print(lst)
```

```python
a, *lst, c = tup
print(tup)
print(a)
print(lst)
print(c)
```

```python
ages = {
    "Alice": 25,
    "Bob": 30,
    "Charlie": 22,
    "Diana": 28,
    "Evan": 35
}

print(ages)
```

```python
mylst = [(k,v) for k,v in ages.items()]
print(mylst)
```

```python
for name,age in mylst:
    print(f"your name is : {name} and you age is {age}")
```
