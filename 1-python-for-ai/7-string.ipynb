# String

```python
txt = "python for machine learning"
print(txt)
print(txt[2:])
print(txt[::2])
```

```python
len(txt)
```

```python
lst_str = txt.split()
print(lst_str)
```

```python
str_txt = ' '.join(lst_str)
print(str_txt, ' ---- ', type(str_txt))
```

```python
str_txt = '-'.join(lst_str)
print(str_txt, ' ---- ', type(str_txt))
```

```python
str_txt = '*'.join(lst_str)
print(str_txt, ' ---- ', type(str_txt))
```

```python
print(str_txt)
print(str_txt.replace('*', " "))
```
