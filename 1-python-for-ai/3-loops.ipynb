# Loops

```python
for i in [1,2,3,4]:
    print(i)
```

```python
for i in range(10):
    print(i**2)
```

```python
print(list(range(1,20,5)))  # range(start, end,step)
```

```python
a = [i for i in range(10) if i%2 == 0]
print(a)
```

```python
b = [i if i%2 == 0 else i / 3 for i in range(10)]
print(b)
```

```python
i = 0
while i < 10:
    i += 1
    if i % 2 == 0:
        continue
    print(i**2)
```

# HOP

```python
for i in range(1,50):
    if i%5 == 0:
        print("HOP")
        continue
    print(i)
```

# Fibonacci

```python
n = 10
n1 = 0
n2 = 1
fibo_lst = [n1, n2]
i = 2
while i < n:
    fibo_lst.append(fibo_lst[i-1] + fibo_lst[i-2])
    i += 1
print(fibo_lst)
```

```python
def fibonacci(n):
    seq = [0, 1]
    for i in range(2, n+1):
        seq.append(seq[i-1] + seq[i-2])
    return seq

print(fibonacci(10))
```

```python
def fib_lst(n):
    res = [0, 1]
    for i in range(2, n+1):
        res.append(res[i-1] + res[i-2])
    return res

fib_lst(10)
```

# count without count

```python
lst = [1,2,3,4,2,3,4,4,3,2,3,3,4,4,4,4,3,2,1]
search = 4
count = 0
for i in lst:
    if i == search:
        count += 1
print(count)
```
